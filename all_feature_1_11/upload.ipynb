```python
import pandas as pd
import numpy as np
```

```python
df_temp = pd.DataFrame()
df_test = pd.read_csv("test_pre.csv")
df_adr = pd.read_csv("all_feature_adr.csv")
df_is_canceled = pd.read_csv("all_feature_is_canceled.csv")
```

```python
df_temp["is_canceled"] = df_is_canceled["is_canceled"]
df_temp["adr"] = df_adr["adr"]
df_temp["total_nights"] = df_test["total_nights"]
df_temp["revenue"] = (df_temp['adr'] * df_temp['total_nights']) * (df_temp['is_canceled'] - 1) * (-1)
df_temp["arrival_date"] = df_test["arrival_date"]
```

```python
df_temp
```

```python
df_revenue_predict = df_temp.groupby(['arrival_date'],as_index=False)['revenue'].agg('sum')
df_revenue_predict['label'] = df_revenue_predict['revenue'].apply(lambda x: 0 if x < 10000 else 
                                            (1 if x < 20000 else
                                            (2 if x < 30000 else
                                            (3 if x < 40000 else
                                            (4 if x < 50000 else
                                            (5 if x < 60000 else
                                            (6 if x < 70000 else
                                            (7 if x < 80000 else
                                            (8 if x < 90000 else 9)))))))))
```

```python
df_revenue_predict.drop(["revenue"], axis=1, inplace=True)
```
